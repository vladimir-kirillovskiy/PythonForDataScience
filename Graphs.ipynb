```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
from pylab import rcParams
import seaborn as sb
import networkx as nx
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

```python
# Creating graph objeck

G = nx.Graph()
nx.draw(G)
```

```python
G.add_node(1)
nx.draw(G)
```

```python
G.add_nodes_from([2,3,4,5,6,7,8,9,12,15,16])
nx.draw(G)
```

```python
G.add_edges_from([(2,4), (2,6), (2,8), (2,12), (2,16), (3,6), (3,9), (3,12), (3,15), (4,8), (4,12), (4,16), (6,12), (8,16)])
nx.draw(G)
```

```python
nx.draw_circular(G)
```

```python
nx.draw_spring(G)
```

```python
nx.draw_circular(G, node_color='bisque', with_labels=True)
```

```python
G.remove_node(1)
nx.draw_circular(G, node_color='bisque', with_labels=True)
```

```python
sum_stats = nx.info(G)
print(sum_stats)
```

```python
print(nx.degree(G))
```

```python
# using graph generators
G = nx.complete_graph(25)
nx.draw(G, node_color='bisque', with_labels=True)
```

```python
G = nx.gnc_graph(7, seed=25)
nx.draw(G, node_color='bisque', with_labels=True)
```

```python
ego_G = nx.ego_graph(G, 3, radius=5)
nx.draw(G, node_color='bisque', with_labels=True)
```
